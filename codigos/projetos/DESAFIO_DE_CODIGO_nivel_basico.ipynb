```python
# Desafio: A Aventura do Explorador

# Entrada
quantidade_passos = int(input())


if quantidade_passos == 0:
    print("Nenhum passo dado na floresta.")

if quantidade_passos > 0 :
    for passos in range(0,quantidade_passos+1):
      if passos==0:
        print(f"Explorador: 1 passo")
      if passos>1:
        print(f"Explorador: {passos} passos")



```

```python
# Desafio: Lista para armazenar os itens

itens = []
total = range(1,4)

for item in total:
  itens.append(input())


# Exibe a lista de itens
print("Lista de itens:")
for item in itens:
    print(f"- {item}")
```
