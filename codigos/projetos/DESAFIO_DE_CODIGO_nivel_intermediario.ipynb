## Desafio
Voc est trabalhando para uma empresa que utiliza extensivamente os servios da AWS, e aps algumas anlises a equipe de segurana identificou que algumas senhas dos usurios no IAM so fracas e podem representar um risco  segurana dos dados da empresa. Para resolver esse problema, foi solicitado que voc desenvolva um programa capaz de analisar as senhas dos usurios e fornecer uma validao de fora com base em critrios predefinidos.

###Requisitos de segurana para a senha:


A senha deve ter no mnimo 8 caracteres.
A senha deve conter pelo menos uma letra maiscula (A-Z).
A senha deve conter pelo menos uma letra minscula (a-z).
A senha deve conter pelo menos um nmero (0-9).
A senha deve conter pelo menos um caractere especial, como !, @, #, $, %, etc.
Saiba mais sobre o Regex em: Java Regular Expressions


## Entrada

A entrada ser uma nica string representando a senha que precisa ser validada.

## Sada

Seu programa deve retornar uma mensagem indicando se a senha fornecida pelo usurio atende aos requisitos de segurana ou no, juntamente com um feedback explicativo sobre os critrios considerados.

```python
def verificar_forca_senha(senha):


  comprimento_minimo = 8
  tem_letra_maiuscula = False
  tem_letra_minuscula = False
  tem_numero = False
  tem_caractere_especial = False

  # Verificando o comprimento da senha
  if len(senha) < comprimento_minimo:
    return f"Sua senha e muito curta. Recomenda-se no minimo {comprimento_minimo} caracteres."

  # TODO: Verifique se a senha contm letras maisculas e minsculas

  tem_maiuscula = any(char.isupper() for char in senha)
  tem_minuscula = any(char.islower() for char in senha)
  tem_numero = any(char.isdigit() for char in senha)
  tem_caractere = any(not(char.isalnum()) for char in senha)

  if not(tem_minuscula and tem_maiuscula and tem_numero and tem_caractere):
      return f"Sua senha nao atende aos requisitos de seguranca."


  # Verificando se a senha contm sequncias comuns

  sequencias_comuns = ["123456", "abcdef"]
  for sequencia in sequencias_comuns:
    if sequencia in senha:
      return "Sua senha contem uma sequencia comum. Tente uma senha mais complexa."

  # Verificando se a senha contm palavras comuns

  palavras_comuns = ["password", "123456", "qwerty"]
  if senha in palavras_comuns:
    return "Sua senha e muito comum. Tente uma senha mais complexa."

  # TODO: Verificar o comprimento mnimo e critrios de validao

  return f"Sua senha atende aos requisitos de seguranca. Parabens!"

# Obtendo a senha do usurio
senha = input().strip() # .strip() remove os espaos na string

# Verificando a fora da senha
resultado = verificar_forca_senha(senha)

# Imprimindo o resultado
print(resultado)
```
